```python
# function name fizz_buzz with parameter num
def fizz_buzz(num):
# if statement 
# check if equality equals to 3 and 5 if yes return FizzBuzz
    if num%3==0 and num%5==0:
        return 'FizzBuzz'
# check if equality equals to 3 if yes return Fizz
    elif num % 3 == 0:
        return 'Fizz'
# check if equality equals to 5 if yes return Buzz
    elif num % 5==0:
        return 'Buzz'
    else:
        return num
# range() starts with 1 and ends with 100
for n in range(1,100):
# prints number range from 1 to 100 using function name fizz_buzz
    print(fizz_buzz(n))
```
